```python
import numpy as np
import pandas as pd
import decisiontree
```

```python
df = pd.read_csv('loan_delay_days.csv')
```

```python
X=df.drop('delay_days', axis=1)
y=np.array(df[['delay_days']])
```

```python
tree = decisiontree.DecisionTreeRegressor(max_depth=3)
```

```python
tree.fit(X.values, y)
```

```python
print(tree.as_json())  # получишь одну строку JSON
```

```python
X.head()
```

```python
print(tree.predict(X.values[:5]))
print(tree._predict_one_sample(X.values[0]))
```
